```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dotted',linewidth=2, label='Kholi')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linewidth=2, marker='D',markersize=10, label='Rohit')
plt.legend(loc='best')
```

```python
plt.bar(np.arange(batsman.shape[0])-0.2,batsman['2015'],width=0.2)
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=0.2)
plt.bar(np.arange(batsman.shape[0])+.20,batsman['2017'],width=0.2)
plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'])
plt.xticks(rotation= 'vertical')
```

# stack bar chart
    sum all values in one graph

```python
plt.bar(batsman['batsman'],batsman['2015'],label='2015')
plt.bar(batsman['batsman'],batsman['2016'],label='2016',bottom=(batsman['2015']))
plt.bar(batsman['batsman'],batsman['2017'],label='2017',bottom=(batsman['2016']+batsman['2015']))
plt.legend()
plt.show()
```
