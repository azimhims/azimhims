```python
import numpy as np


IMDB = previous['IMDB'].values
RottenTomatoes = previous['RottenTomatoes'].values
Matacritic = previous['Metacritic'].values
RT_norm = previous['RT_norm'].values
Fandango_Star = previous['Fandango_Stars'].values


def scale_to_range(arr, target_min, target_max):
    min_val = np.min(arr)
    max_val = np.max(arr)
    scaled_arr = ((arr - min_val) / (max_val - min_val)) * (target_max - target_min) + target_min
    return scaled_arr


IMDB_scaled = np.round(scale_to_range(IMDB, 1, 10),decimals = 2)
RottenTomatoes_scaled = np.round(scale_to_range(RottenTomatoes, 1, 10),decimals=2)
Matacritic_scaled = np.round(scale_to_range(Matacritic, 1, 10),decimals=2)
RT_norm_scaled = np.round(scale_to_range(RT_norm, 1, 10),decimals=2)
Fandango_Star_scaled = np.round(scale_to_range(Fandango_Star, 1, 10),decimals=2)


previous['IMDB_Scaled'] = np.round(scale_to_range(IMDB, 1, 10),decimals = 2)
previous['RottenTomatoes_scaled'] = np.round(scale_to_range(RottenTomatoes, 1, 10),decimals=2)
previous['Matacritic_scaled'] = np.round(scale_to_range(Matacritic, 1, 10),decimals=2)
previous['RT_norm_scaled'] =np.round(scale_to_range(RT_norm, 1, 10),decimals=2)
previous['Fandango_Star_scaled'] = np.round(scale_to_range(Fandango_Star, 1, 10),decimals=2)


print("IMDB_scaled:", IMDB_scaled)
print("RottenTomatoes_scaled:", RottenTomatoes_scaled)
print("Matacritic_scaled:", Matacritic_scaled)
print("RT_norm_scaled:", RT_norm_scaled)
print("Fandango_Star_scaled:", Fandango_Star_scaled)
```
