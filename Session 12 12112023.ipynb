```python
eu_country = ''' Austria, Belgium, Bulgaria, Croatia, Cyprus, Czechia, Denmark, Estonia, Finland, France, Germany, Greece, Hungary, Ireland, Italy, Latvia, Lithuania, Luxembourg, Malta, Netherlands, Poland, Portugal, Romania, Slovakia, Slovenia, Spain, Sweden'''
```

```python
eu = eu_country.split(",").lstrip()
```

```python
eu_list = []
for i in eu:
    i = i.strip()
    eu_list.append(i)
```

```python
eu_list
```

```python
euro_death = who[who['Country'].isin(eu_list)]
```

```python
euro_death['Date_reported'] = pd.to_datetime(euro_death['Date_reported']).copy()
```

```python
euro_death
```

```python
df = euro_death.groupby('Date_reported')['Cumulative_cases','New_cases'].sum()
```

```python
new_df = pd.DataFrame(df)
```
