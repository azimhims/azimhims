```python
import pandas as pd
import numpy as np
import seaborn as  sns
import matplotlib.pyplot as plt
```

```python
dataset_path = ("D:/Azeem Data/D Drive/kaggle/house price advance regression/train - Copy.csv")
df = pd.read_csv(dataset_path)
```

```python
df.head()
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
nul_col = df.isnull().sum()/ df.shape[0]*100
```

```python
missing_col = nul_col[nul_col >20].keys()
```

```python
missing_col
```

```python
df2_drop_col = df.drop(columns=missing_col)
df2_drop_col.shape
```

```python
df3_num = df2_drop_col.select_dtypes(include=['int64','float64'])
df3_num
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df3_num.isnull())
```

```python
#that show the column where identify the mission value if any.
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()/df3_num.shape[0]*100
```

```python
#with loop we identify the columns name where mission value avaliable if any 
missing_num_val = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_val
```

```python
#now before fill the missing value, we need to check either missing value fill by mean or median, use simple distribution

plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20, kde_kws={'linewidth':5,'color':'red'})
```

```python
df4_num_means = df3_num.fillna(df3_num.mean())
```

```python
df4_num.isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws= {'linewidth':2,'color':'Red'})
    sns.distplot(df4_num[var], bins=20,kde_kws= {'linewidth':2, 'color':"Green"})
    plt.legend()
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_val):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,hist = False, kde_kws={'linewidth':2,'color':'Red'},label="Orignal")
    sns.distplot(df4_num_means[var],bins=20,hist = False, kde_kws={'linewidth':2,'color':'black'},label='Mean')
    sns.distplot(df5_num_median[var],bins=20,hist = False, kde_kws={'linewidth':2, 'color':'green'},label='Median')
    plt.legend()
```

```python
#identify outlier using boxplot

for i, var in enumerate(missing_num_val):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_num[var])
```

```python
#CONCATENATE ROW WISE OF MISSING VALUES AMOUNT ORIGNAL, MEANS, AND MEDIAN
pd_concat = pd.concat([df3_num[missing_num_val],df4_num_means[missing_num_val],df5_num_median[missing_num_val]],axis=1)
```

```python
#THAT SHOWN THE NULL VALUE IN ORIGNAL VALUE AS WELL AS MEANS AND MEDIAN
pd_concat[pd_concat.isnull().any(axis=1)]
```
